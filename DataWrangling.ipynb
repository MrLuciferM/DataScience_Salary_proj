```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
path = 'https://raw.githubusercontent.com/PlayingNumbers/ds_salary_proj/data_cleaning/glassdoor_jobs.csv'
```

```python
df = pd.read_csv(path,index_col=False)
df
```

```python
# deleting rows without salary
df = df[df['Salary Estimate'] != '-1']
df['Salary Estimate']
```

```python
# separating hourly salary
df.loc[:,'hourly_sal'] = df['Salary Estimate'].apply(lambda x: 1 if 'PER HOUR' in x.upper() else 0)
df.hourly_sal
```

```python
# separating Employer provided salary
df.loc[:,'emp_provided_sal'] = df['Salary Estimate'].apply(lambda x: 1 if 'EMPLOYER PROVIDED SALARY:' in x.upper() else 0)
df.emp_provided_sal
```

```python
# cleaning salary
salary = df['Salary Estimate'].apply(lambda x: x.split('(')[0])
salary
```

```python
# removing $ and K
salary = salary.apply(lambda x: x.replace('$','').replace('K',''))
salary
```

```python
# removing per hour and employer provided salary
salary = salary.apply(lambda x: x.upper().replace('PER HOUR','').replace('EMPLOYER PROVIDED SALARY:',''))
salary
```

```python
# creating minimum, maximum and average salary
df.loc[:,'min_sal'] = salary.apply(lambda x: int(x.split('-')[0]))
df.loc[:,'max_sal'] = salary.apply(lambda x: int(x.split('-')[1]))
df.loc[:,'avg_sal'] = (df['min_sal'] + df['max_sal'])/2
```

```python
# converting hourly salary to annual
# all salaries are in $1000s 
df.loc[:,'min_sal'] = df.apply(lambda x: x['min_sal']*2 if x['hourly_sal'] == 1 else x['min_sal'], axis=1)
df.loc[:,'max_sal'] = df.apply(lambda x: x['max_sal']*2 if x['hourly_sal'] == 1 else x['max_sal'], axis=1)
df.loc[:,'avg_sal'] = df.apply(lambda x: x['avg_sal']*2 if x['hourly_sal'] == 1 else x['avg_sal'], axis=1)
```

```python
df[df['hourly_sal'] == 1][['hourly_sal','min_sal','avg_sal','max_sal']]
```

```python
df.dtypes
```

```python
# removing Multiple lines from company name
df.loc[:,'Company Name'] = df['Company Name'].apply(lambda x: x.split('\n')[0])
df['Company Name']
```

```python
# finding state of the job
df.loc[:,'job_state'] = df['Location'].apply(lambda x: x.split(',')[1])
df.job_state
```

```python
df.job_state.value_counts()
```

```python
# fixing Los Angeles
df.loc[:,'job_state'] = df.job_state.apply(lambda x: x.strip() if x.strip().lower() != 'los angeles' else 'CA')
df.job_state.value_counts()
```

```python
# finding jobs in same state as company headquarters 
df.loc[:,'same_state'] = df.apply(lambda x: 1 if x.Location == x.Headquarters else 0, axis=1)
df.same_state
```

```python
# finding how old is the company
df.loc[:,'company_age'] = df.Founded.apply(lambda x: x if x<0 else 2020-x)
df.company_age.value_counts()
```

```python
# Technology requirements from job description 

# python
df.loc[:,'python'] = df['Job Description'].apply(lambda x: 1 if 'PYTHON' in x.upper() else 0)
print("python\n",df.python.value_counts())
# r language
df.loc[:,'r_lang'] = df['Job Description'].apply(lambda x: 1 if ('R STUDIO' in x.upper() or 'R LANGUAGE' in x.upper() or 'R PROGRAMMING' in x.upper()) else 0)
print("r_lang\n",df.r_lang.value_counts())
# excel
df.loc[:,'excel'] = df['Job Description'].apply(lambda x: 1 if 'EXCEL' in x.upper() else 0)
print("excel\n",df.excel.value_counts())
# anaconda
df.loc[:,'anaconda'] = df['Job Description'].apply(lambda x: 1 if 'ANACONDA' in x.upper() else 0)
print("anaconda\n",df.anaconda.value_counts())
# rapidminer
df.loc[:,'rapid'] = df['Job Description'].apply(lambda x: 1 if 'RAPIDMINER' in x.upper() else 0)
print("rapid\n",df.rapid.value_counts())
# sql
df.loc[:,'sql'] = df['Job Description'].apply(lambda x: 1 if 'SQL' in x.upper() else 0)
print("sql\n",df.sql.value_counts())
# tensorflow
df.loc[:,'tensorflow'] = df['Job Description'].apply(lambda x: 1 if 'TENSORFLOW' in x.upper() else 0)
print("tensor\n",df.tensorflow.value_counts())
# tableau
df.loc[:,'tableau'] = df['Job Description'].apply(lambda x: 1 if 'TABLEAU' in x.upper() else 0)
print("tableau\n",df.tableau.value_counts())
# spark
df.loc[:,'spark'] = df['Job Description'].apply(lambda x: 1 if 'SPARK' in x.upper() else 0)
print("spark\n",df.spark.value_counts())
# AWS
df.loc[:,'aws'] = df['Job Description'].apply(lambda x: 1 if 'AWS' in x.upper() else 0)
print("AWS\n",df.aws.value_counts())
```

```python
def simplifyJob(title):
    '''
    This function will simplify Job title from the data frame and return simplified title
    '''
    if'data scientist' in title.lower():
        return 'Data Scientist'
    elif 'data engineer' in title.lower():
        return 'Data Engineer'
    elif 'machine learning' in title.lower():
        return 'MLE'
    elif 'analyst' in title.lower():
        return 'Analyst'
    elif 'manager' in title.lower():
        return 'Manager'
    elif 'director' in title.lower():
        return 'Director'
    else:
        return 'NA'

def seniority(title):
    '''
    this function will find the Seniority level of the job
    '''
    if 'sr' in title.lower() or 'senior' in title.lower() or 'lead' in title.lower() or 'principal' in title.lower():
        return 'Sr'
    elif 'jr' in title.lower() or 'junior' in title.lower():
        return 'Jr'
    else:
        return 'NA'
```

```python
# simplifying job title
df.loc[:,'job_simplified'] = df['Job Title'].apply(simplifyJob)
df.job_simplified.value_counts()
```

```python
# seniority level of the job
df.loc[:,'seniority'] = df['Job Title'].apply(seniority)
df.seniority.value_counts()
```

```python
# job description length
df.loc[:,'job_description_len'] = df['Job Description'].apply(lambda x: len(x))
df.job_description_len
```

```python
# number of competitors of the company
df.loc[:,'num_of_competetors'] = df.Competitors.apply(lambda x: len(x.split(',')) if x != '-1' else 0)
df.num_of_competetors
```
