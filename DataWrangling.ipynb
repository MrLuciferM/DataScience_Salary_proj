```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```
path = 'https://raw.githubusercontent.com/PlayingNumbers/ds_salary_proj/data_cleaning/glassdoor_jobs.csv'
```

```
df = pd.read_csv(path)
df
```

```
df = df[df['Salary Estimate'] != '-1']
df
```

```
# separating hourly salary
df['hourly_sal'] = df['Salary Estimate'].apply(lambda x: 1 if 'PER HOUR' in x.upper() else 0)
df
```

```
df['emp_provided_sal'] = df['Salary Estimate'].apply(lambda x: 1 if 'EMPLOYER PROVIDED SALARY:' in x.upper() else 0)
df
```

```
salary = df['Salary Estimate'].apply(lambda x: x.split('(')[0])
salary
```

```
salary = salary.apply(lambda x: x.replace('$','').replace('K',''))
salary
```

```
salary = salary.apply(lambda x: x.upper().replace('PER HOUR','').replace('EMPLOYER PROVIDED SALARY:',''))
salary
```

```
df['min_sal'] = salary.apply(lambda x: int(x.split('-')[0]))
df['max_sal'] = salary.apply(lambda x: int(x.split('-')[1]))
df['avg_sal'] = (df['min_sal'] + df['max_sal'])/2
df.dtypes
```

```
df['Company Name'] = df['Company Name'].apply(lambda x: x.split('\n')[0])
df
```

```
df['job_state'] = df['Location'].apply(lambda x: x.split(',')[1])
df
```

```
df['job_state'].value_counts()
```

```
df['same_state'] = df.apply(lambda x: 1 if x.Location == x.Headquarters else 0, axis=1)
df
```

```
df['company_age'] = df.Founded.apply(lambda x: x if x<0 else 2020-x)
df
```

```
# python
df['python'] = df['Job Description'].apply(lambda x: 1 if 'PYTHON' in x.upper() else 0)
print("python\n",df.python.value_counts())
# r language
df['r_lang'] = df['Job Description'].apply(lambda x: 1 if ('R STUDIO' in x.upper() or 'R LANGUAGE' in x.upper() or 'R PROGRAMMING' in x.upper()) else 0)
print("r_lang\n",df.r_lang.value_counts())
# excel
df['excel'] = df['Job Description'].apply(lambda x: 1 if 'EXCEL' in x.upper() else 0)
print("excel\n",df.excel.value_counts())
# anaconda
df['anaconda'] = df['Job Description'].apply(lambda x: 1 if 'ANACONDA' in x.upper() else 0)
print("anaconda\n",df.anaconda.value_counts())
# rapidminer
df['rapid'] = df['Job Description'].apply(lambda x: 1 if 'RAPIDMINER' in x.upper() else 0)
print("rapid\n",df.rapid.value_counts())
# sql
df['sql'] = df['Job Description'].apply(lambda x: 1 if 'SQL' in x.upper() else 0)
print("sql\n",df.sql.value_counts())
# tensorflow
df['tensorflow'] = df['Job Description'].apply(lambda x: 1 if 'TENSORFLOW' in x.upper() else 0)
print("tensor\n",df.tensorflow.value_counts())
# tableau
df['tableau'] = df['Job Description'].apply(lambda x: 1 if 'TABLEAU' in x.upper() else 0)
print("tableau\n",df.tableau.value_counts())
# spark
df['spark'] = df['Job Description'].apply(lambda x: 1 if 'SPARK' in x.upper() else 0)
print("spark\n",df.spark.value_counts())
# AWS
df['aws'] = df['Job Description'].apply(lambda x: 1 if 'AWS' in x.upper() else 0)
print("AWS\n",df.aws.value_counts())
```

```
df = df.drop('Unnamed: 0', axis=1)
df_cleaned=df.reset_index()
df_cleaned.drop('index',axis=1, inplace=True)
df_cleaned
```

```
df_cleaned.columns
```

```
df_cleaned['Job Title'].value_counts()
```

```
def simplifyJob(title):
  if 'data scientist' in title.lower():
    return 'Data Scientist'
  elif 'data engineer' in title.lower():
    return 'Data Engineer'
  elif 'machine learning' in title.lower():
    return 'MLE'
  elif 'analyst' in title.lower():
    return 'Analyst'
  elif 'manager' in title.lower():
    return 'Manager'
  elif 'director' in title.lower():
    return 'Director'
  else:
    return 'NA'

def seniority(title):
  if 'sr' in title.lower() or 'senior' in title.lower() or 'lead' in title.lower() or 'principal' in title.lower():
    return 'Sr'
  elif 'jr' in title.lower() or 'junior' in title.lower():
    return 'Jr'
  else:
    return 'NA'
```

```
df_cleaned['job_simplified'] = df_cleaned['Job Title'].apply(simplifyJob)
df_cleaned.job_simplified.value_counts()
```

```
df_cleaned['seniority'] = df_cleaned['Job Title'].apply(seniority)
df_cleaned.seniority.value_counts()
```

```
df_cleaned['job_state'].value_counts()
```

```
df_cleaned['job_state'] = df_cleaned['job_state'].apply(lambda x: x.strip() if x.strip().lower() != 'los angeles' else 'CA')
df_cleaned['job_state'].value_counts()
```

```
# job description length
df_cleaned['job_description_len'] = df_cleaned['Job Description'].apply(lambda x: len(x))
df_cleaned.job_description_len
```

```
# number of competitors
df_cleaned['num_of_competetors'] = df_cleaned.Competitors.apply(lambda x: len(x.split(',')) if x != '-1' else 0)
df_cleaned.num_of_competetors
```

```
# converting hourly salary to annual
# all salaries are in $1000s 
df_cleaned['min_sal'] = df_cleaned.apply(lambda x: x['min_sal']*2 if x['hourly_sal'] == 1 else x['min_sal'], axis=1)
df_cleaned['max_sal'] = df_cleaned.apply(lambda x: x['max_sal']*2 if x['hourly_sal'] == 1 else x['max_sal'], axis=1)
df_cleaned['avg_sal'] = df_cleaned.apply(lambda x: x['avg_sal']*2 if x['hourly_sal'] == 1 else x['avg_sal'], axis=1)
```
